```python
# Import Essential Libraries 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("C:/Users/Sistech Computer/OneDrive/Desktop/sandeep/churn-prediction-retentional-analysis/data/model_output/churn_predictions_full.csv")
df.head()
```

```python
fig=px.histogram(df, x='churn_probability', nbins=30, title='Distribution of Churn Probability', color='Churn')
fig.show()
```

## Insight
###### Model ka probability output do classes ke beech achhi tarah separate hai — actual non-churners ka average 
###### predicted probability sirf 0.258 hai (median 0.18), jabki actual churners ka average 0.596 hai (median 0.62).
###### Ye clear separation dikhata hai ki model confidently distinguish kar pa raha hai, random guessing nahi kar 
###### raha.

```python
fig=px.density_heatmap(df, x='InternetService',
                       y='Churn',
                       title='Density Heatmap of Internet Service and Churn'
                       )
fig.show()
```

## Insight
###### Fiber optic customers ka churn rate 41.1% hai, DSL ka 20.0%, aur no-internet customers ka sirf 8.0% — matlab
###### Fiber optic users 5x zyada churn karte hain no-internet segment se. Ye Phase 5 ke LR coefficient 
###### (InternetService_Fiber optic: +0.353, sabse strong positive driver) ko directly confirm karta hai.

```python
fig=px.density_heatmap(df, x='Contract', y='Churn', title='Density Heatmap of Contract and Churn')
fig.show()
```
